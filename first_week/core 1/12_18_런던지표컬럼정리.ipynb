### 스마트 미더 (AMI) 원천 데이터 확보 후 CSV 구조 확인. 이후에 데이터가 코에 1에 적합한지를 검증함 

```python
import pandas as pd
df = pd.read_csv("../data_csv/런던계절별사용량.csv")
```

#### 컬럼 정리 

```python
df = df.rename(columns={"LCLid": "household_id", "DateTime": "timestamp", "KWHhh": "consumption"})
```

```python
df["timestamp"] = pd.to_datetime(df["timestamp"])
```

```python
df = df[df["household_id"] == "MAC000001"].sort_values("timestamp")
```

#### 인덱스 컬럼은 필요 없어서 drop 

```python
df = df.drop(columns=["Index"])
df = df.drop(columns=["Tariff"])
```

#### timestamp column을 index 로 설정 

```python
df = df.set_index("timestamp")
```

```python
df.info()
```

```python
df.index
```

#### 날씨csv와 합침 

```python
weather = pd.read_csv("./data_csv/런던날씨.csv")
```

```python
weather["time"] = pd.to_datetime(weather["time"])
weather = weather.set_index("time")
```

```python
df_merged = df.join(weather, how="left")
```

```python
df_merged.info()
```

#### csv 파일 생성

```python
df_merged.to_csv(".//data_csv/amiwea.csv")
```
