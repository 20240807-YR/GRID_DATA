```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("../data_csv/amiwea.csv", parse_dates=["timestamp"])
df = df.set_index("timestamp")
```

#### 산점도: 기온 VS 전력 소비 
##### 외생변수 (기온)와 부하 간 관계 시각화 및 "날씨를 왜 붙였는지를 설명 

```python
plt.figure(figsize=(6, 5))
plt.scatter(df["temperature"], df["consumption"], alpha=0.3, s=10)
plt.xlabel("Temperature (°C)")
plt.ylabel("Electricity Consumption (kWh)")
plt.title("Temperature vs Electricity Consumption")
plt.tight_layout()
plt.show()
```

#### 시간축 겹쳐 보기
##### 동일한 시간대에서의 동조  역상 관계 관찰, EMS 수요예측으로 자연스럽게 연결 가능 

```python
fig, ax1 = plt.subplots(figsize=(15, 4))

ax1.plot(df.index, df["consumption"], color="tab:blue", linewidth=0.8)
ax1.set_ylabel("Consumption (kWh)", color="tab:blue")

ax2 = ax1.twinx()
ax2.plot(df.index, df["temperature"], color="tab:red", alpha=0.5)
ax2.set_ylabel("Temperature (°C)", color="tab:red")

plt.title("Electricity Consumption and Temperature Over Time")
plt.tight_layout()
plt.show()
```

“전력 소비량과 기온 간의 관계를 분석한 결과,
기온 변화에 따라 소비 패턴이 달라지는 경향이 관찰되었으며,
이는 외생 변수(날씨)를 포함한 부하 예측의 필요성을 시사한다.”
